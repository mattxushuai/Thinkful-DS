```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv(r"C:\Users\M&D\Desktop\who_suicide_statistics.csv\who_suicide_statistics.csv")
print(df)
```

```python
df[df['suicides_no'] > 0]
```

```python
df_by_year = df.groupby('year').aggregate(np.sum)
print(df_by_year)
```

```python
plt.plot(df_by_year['suicides_no'], color = 'red')
plt.ylabel('Suicide number')
plt.title('Suicide number 1979 - 2016')
plt.show()
```

This plot shows the relationship between total suicide numbers and year. There's a upward trend from 1979 to 2003 and slightly downward trend from that time on.

```python
df_by_year['suicide_rate'] = df_by_year['suicides_no'] * 1000000 / df_by_year['population']
plt.plot(df_by_year['suicide_rate'], color = 'blue')
plt.ylabel('suicide per million people')
plt.title('Suicide rate 1979 - 2016 (per million people)')
plt.show()
```

This plot shows the suicide rate (suicide number per million people). The graph shows that the number fluctuates but overall is stable before 1995, and gradually decline from that year on.

```python
df_by_gender = df.groupby('sex').aggregate(np.sum)
print(df_by_gender)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1, 2, 1)
plt.plot(df_by_year['suicides_no'], color = 'red')
plt.ylabel('Suicide number')
plt.title('Suicide number 1979 - 2016')

plt.subplot(1, 2, 2)
plt.plot(df_by_year['suicide_rate'], color = 'blue')
plt.ylabel('suicide per million people')
plt.title('Suicide rate 1979 - 2016 (per million people)')
plt.show()
```

This is a subplot of the two plots above.

```python
df_by_gender.loc[:, 'suicides_no'].plot(kind='bar')
```

This is the overall suicide numbers of female and male, respectively, from 1979 to 2016. Over all, male suicide number triples that of female.

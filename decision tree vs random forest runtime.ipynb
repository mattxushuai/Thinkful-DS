```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import time
```

```python
y2015 = pd.read_csv('LoanStats3d.csv',skipinitialspace=True, header=1)
```

```python
y2015 = y2015[:-2]
y2015.dropna(axis=1, thresh=20000,inplace=True)
```

```python
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

y2015.drop(['emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state'], 1, inplace=True)
```

```python
pd.get_dummies(y2015)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

start_time = time.time()
rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)
```

```python
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
from sklearn import tree
start_time = time.time()
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=1,
    max_depth=4  
)

cross_val_score(decision_tree, X, Y, cv=10)
print("--- %s seconds ---" % (time.time() - start_time))
```

## random forest with no PCA took 184 seconds while decision tree took 30 seconds.

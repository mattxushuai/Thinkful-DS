```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head(2)
```

## Choose one variable and plot that variable four different ways.

```python
sns.distplot(tips.total_bill, color = 'green', kde = False)
plt.show()
```

```python
sns.boxplot(tips.total_bill, hue="sex", data=tips)
plt.show()
```

```python
sns.pointplot(tips.total_bill)
plt.show()
```

```python
sns.kdeplot(tips.total_bill, color = 'purple', lw = 2, ls = '--')
plt.show()
```

## Choose two continuous variables, and plot them three different ways.

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.lmplot(x="size", y="tip", data=tips,x_jitter=0.1)
plt.show()
```

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips)
plt.show()
```

## Choose one continuous variable and one categorical variable, and plot them six different ways.

```python
sns.barplot(x='sex',y='total_bill',data=tips, palette='rainbow')
plt.show()
```

```python
sns.boxplot(x="sex", y="total_bill", data=tips,palette='rainbow')
plt.show()
```

```python
sns.violinplot(x="sex", y="total_bill", data=tips,palette='coolwarm')
plt.show()
```

```python
sns.stripplot(x="sex", y="total_bill", data=tips,jitter=True)
plt.show()
```

```python
sns.swarmplot(x="sex", y="total_bill", data=tips, hue="smoker")
plt.show()
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='bar')
plt.show()
```

## Challenge

### Probability distributions

```python
sns.pairplot(tips, hue='sex',palette='coolwarm')
plt.show()
```

### Bivariate relationships

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex', color='green')
plt.show()
```

### Whether the distributions or the relationships vary across groups

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex',palette='coolwarm')
plt.show()
```

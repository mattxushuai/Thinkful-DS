```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exponential = np.random.exponential(2, 100)
plt.hist(exponential)
plt.axvline(exponential.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponential.mean() + exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponential.mean() - exponential.std(), color='b', linestyle='dashed', linewidth=2)
plt.show()
```

Descriptive statistics not useful

```python
norm1 = np.random.normal(5,.5, 100)
norm2 = np.random.normal(10,1, 100)
combine = norm1+norm2

mean = np.mean(combine)
sd = np.std(combine)

plt.hist(combine)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
print('combination of two normal distribution with different centers should show some skew.')
```

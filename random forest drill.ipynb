```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
y2015 = pd.read_csv('LoanStats3d.csv',skipinitialspace=True, header=1)
```

```python
y2015 = y2015[:-2]
y2015.dropna(axis=1, thresh=20000,inplace=True)
```

```python
y2015.tail()
```

```python
y2015.info()
```

```python
categorical = y2015.select_dtypes(include=['object'])
for i in categorical:
    column = categorical[i]
    print(i)
    print(column.nunique())
```

```python
y2015['int_rate'] = pd.to_numeric(y2015['int_rate'].str.strip('%'), errors='coerce')

y2015.drop(['emp_title', 'zip_code', 'earliest_cr_line', 'revol_util',
            'sub_grade', 'addr_state'], 1, inplace=True)
```

```python
pd.get_dummies(y2015)
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()
X = y2015.drop('loan_status', 1)
Y = y2015['loan_status']
X = pd.get_dummies(X)
X = X.dropna(axis=1)

cross_val_score(rfc, X, Y, cv=10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents)
principalDf.head()
```

```python
X_pca_10 = principalDf
Y = y2015['loan_status']
X_pca_10 = pd.get_dummies(X_pca_10)
X_pca_10 = X_pca_10.dropna(axis=1)

cross_val_score(rfc, X_pca_10, Y, cv=10)
```

```python
pca = PCA(n_components=40)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
X_pca_40 = principalDf
Y = y2015['loan_status']
X_pca_40 = pd.get_dummies(X_pca_40)
X_pca_40 = X_pca_40.dropna(axis=1)

cross_val_score(rfc, X_pca_40, Y, cv=10)
```

```python
pca = PCA(n_components=30)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
X_pca_30 = principalDf
Y = y2015['loan_status']
X_pca_30 = pd.get_dummies(X_pca_30)
X_pca_30 = X_pca_30.dropna(axis=1)

cross_val_score(rfc, X_pca_30, Y, cv=10)
```

```python
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
X_pca_20 = principalDf
Y = y2015['loan_status']
X_pca_20 = pd.get_dummies(X_pca_20)
X_pca_20 = X_pca_20.dropna(axis=1)

cross_val_score(rfc, X_pca_20, Y, cv=10)
```

```python
pca = PCA(n_components=10)
principalComponents = pca.fit_transform(X)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
X_pca_10 = principalDf
Y = y2015['loan_status']
X_pca_10 = pd.get_dummies(X_pca_10)
X_pca_10 = X_pca_10.dropna(axis=1)

cross_val_score(rfc, X_pca_10, Y, cv=10)
```
